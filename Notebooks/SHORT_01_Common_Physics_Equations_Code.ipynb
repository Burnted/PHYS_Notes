# Short Notebook 1: Common Physics Equations (Code Implementations)



```python
import numpy as np
from scipy.constants import G, g, epsilon_0, mu_0

```

```python
# Kinematics: x = x0 + v0 t + 0.5 a t^2
def position_constant_accel(t, x0, v0, a):
    return x0 + v0*t + 0.5*a*t**2

print(position_constant_accel(2.0, 0.0, 3.0, -9.81))

```

```python
# Coulomb force magnitude
def coulomb_force(q1, q2, r):
    return abs(q1*q2)/(4*np.pi*epsilon_0*r**2)

print(coulomb_force(1e-6, 2e-6, 0.1))

```

```python
# Newtonian gravity force magnitude
def gravity_force(m1, m2, r):
    return G*m1*m2/r**2

print(gravity_force(1.0, 5.0, 2.0))

```

```python
# Harmonic oscillator period
def sho_period(m, k):
    return 2*np.pi*np.sqrt(m/k)

print(sho_period(0.2, 10.0))

```

```python
# Relativistic gamma
def gamma(v, c0=299792458.0):
    beta2 = (v/c0)**2
    return 1/np.sqrt(1-beta2)

print(gamma(0.8*299792458.0))

```
