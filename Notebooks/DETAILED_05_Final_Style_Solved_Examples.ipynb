```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad, solve_ivp
from scipy.optimize import root, minimize
from scipy.interpolate import interp1d
from numpy.polynomial.chebyshev import chebinterpolate, chebval

```

## Problem 1: Root finding in a transcendental equation
Solve `cos(x)=x` and verify residual.



```python
f = lambda x: np.cos(x)-x
sol = root(lambda z: [f(z[0])], [0.5])
xstar = sol.x[0]
print('solution x* =', xstar)
print('residual =', f(xstar))

```

## Problem 2: Numerical integral and uncertainty
Compute \(\int_0^3 e^{-x} \sin(2x) dx\).



```python
I, err = quad(lambda x: np.exp(-x)*np.sin(2*x), 0, 3)
print('integral =', I)
print('estimated error =', err)

```

## Problem 3: ODE dynamics with `solve_ivp`
Damped oscillator with given initial conditions.



```python
def rhs(t, u, zeta=0.12, w0=3.0):
    x, v = u
    return [v, -2*zeta*w0*v - w0**2*x]

t = np.linspace(0, 12, 600)
sol3 = solve_ivp(rhs, [t[0], t[-1]], [1.0, 0.0], t_eval=t)

plt.plot(sol3.t, sol3.y[0])
plt.title('Damped oscillator response')
plt.xlabel('t'); plt.ylabel('x(t)')
plt.show()

```

## Problem 4: Interpolation and query
Given sampled data, estimate values at unsampled points.



```python
x = np.array([0,1,2,3,4,5], dtype=float)
y = np.array([0.0,0.9,0.2,-0.7,-0.5,0.4])
fint = interp1d(x, y, kind='cubic')
q = np.array([0.5, 2.5, 4.2])
print('queries:', q)
print('interpolated:', fint(q))

```

## Problem 5: Chebyshev approximation
Approximate a smooth function and evaluate quickly.



```python
g = lambda x: np.exp(x)
coef = chebinterpolate(g, deg=8)
xs = np.linspace(-1, 1, 300)
err = np.max(np.abs(chebval(xs, coef) - g(xs)))
print('max error on [-1,1] =', err)

```

## Problem 6: Parameter fit by minimization
Fit a model by minimizing sum of squared residuals.



```python
rng = np.random.default_rng(123)
x = np.linspace(0, 4, 40)
y = 1.7*np.exp(-0.8*x) + 0.2 + rng.normal(0, 0.03, x.size)

model = lambda p, x: p[0]*np.exp(-p[1]*x) + p[2]
obj = lambda p: np.sum((y - model(p, x))**2)
res = minimize(obj, x0=[1.0, 1.0, 0.0])
print('best-fit parameters [A,k,C]=', res.x)

plt.scatter(x, y, s=15, label='data')
plt.plot(x, model(res.x, x), 'r', label='fit')
plt.legend(); plt.title('Exponential fit via minimize')
plt.show()

```

## Problem 7: Mixed-method exam-style mini set
- Integral + interpolation consistency check
- ODE endpoint condition check
- Fit quality comparison via residual norm



```python
# mini check 1: integral of interpolated curve over data range
xx = np.linspace(x.min(), x.max(), 400)
yy = model(res.x, xx)
print('trapz area under fit =', np.trapezoid(yy, xx))

# mini check 2: endpoint from ODE
print('final state [x,v]=', sol3.y[:, -1])

# mini check 3: residual norm
r = y - model(res.x, x)
print('||residual||2 =', np.linalg.norm(r))

```
